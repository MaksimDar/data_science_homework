3. Створіть масив розміром 5x5, заповніть його випадковими цілими числами в діапазоні від 1 до 10 та виведіть його значення.

```python
# start_index,matrix_name, minimal range and maximum range) 
def create_matrix(index, matrix, min_range, max_range):
    while index <= len(matrix) - 1:
        i = 0
        for i in range(len(matrix)):
            matrix[index][i] = np.random.randint(min_range,max_range+1)
        index+=1
    return matrix
```

```python
matrix_5 = np.zeros((5,5), dtype=int)
min_range = 1
max_range = 10
start_index = 0
matrix_5 = create_matrix(start_index, matrix_5, min_range,max_range)
print(matrix_5)
```

4. Створіть наступний шаблон без жорсткого кодування(Hard coding) . Використовуйте лише функції Numpy і наведений нижче масив введення Input:a = np.array([1,2,3])

Output:array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])

```python
a = np.array([1,2,3])
print('a matrix')
print(a)

part_1 = np.repeat(a,len(a))
part_2 = np.tile(a,len(a))
result_m = np.concatenate((part_1,part_2))
print(result_m)
```

10. Знайдіть повторювані записи (від 2-го випадку повторення і далі) у заданому масиві і позначте їх як True. Перше повторення має бути False.
np.random.seed(100) a = np.random.randint(0, 5, 10)
[0 0 3 0 2 4 2 2 2 2] #Output [False True False True False False True True True True]

```python
np.random.seed(100)
a = np.random.randint(0, 5, 10)
_, idx = np.unique(a, return_index=True)
mask = np.ones(len(a), dtype=bool)
mask[idx] = False
print('Output: ', mask)
```
