Дане домашнє завдання буде повністю пов'язане з лінійною регресією та її реалізацією. Отож розіб'ємо наше домашнє завдання на декілька частин:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

напишіть функцію гіпотези лінійної регресії у векторному вигляді;

```python
def linear_regression_hypothesis(x,w):
    model = x.dot(w)
    return model
```

створіть функцію для обчислення функції втрат у векторному вигляді;

```python
def error_function(x,y,w):
    y_hat = linear_regression_hypothesis(x,w)
    mean_squarred_error = (1/(2*len(y))) * np.sum((y_hat - y)**2)
    return mean_squarred_error
```

реалізуйте один крок градієнтного спуску;

```python
# alpha == learning rate
def gradient_descent(x,y,w,alpha = 0.0001):
    m = len(y)
    y_hat = linear_regression_hypothesis(x,w)
    gradient = (1/m) * x.T.dot(y_hat - y)
    w_new = w - alpha * gradient
    return w_new
```

знайдіть найкращі параметри w для датасету використовуючи написані вами функції, прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень;

```python
df = pd.read_csv("Housing.csv")
```

```python
X_raw = df[["area","bathrooms","bedrooms"]].values
y = df["price"].values

X_raw_n = (X_raw - X_raw.mean(axis=0)) / (X_raw.max(axis=0) - X_raw.min(axis=0))
y_n = (y - y.mean()) / (y.max() - y.min())
bias = np.ones([len(X_raw_n),1])

X = np.concatenate([bias,X_raw_n],axis=1)

weights_g = np.zeros(X.shape[1])
for i in range(500000):
    weights_g = gradient_descent(X,y_n,weights_g)
mse_g = error_function(X,y_n,weights_g)
# print(f"Best parameters for weights: {weights_g}")
# print(f"Mean squarred error: {mse_g}")
```

знайдіть ці ж параметри за допомогою аналітичного рішення;

```python
weights_a = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_n)
mse_a = error_function(X, y_n, weights_a)
# print(f"Best parameters for weights using analytical method {weights_a}")
# print(f"Mean squarred error using analytical weights: {mse_analytical}")
```

для перевірки спрогнозованих значень, використайте LinearRegression з бібліотеки scikit-learn та порівняйте результати.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True).fit(X_raw_n,y_n)

pred_sk = model.predict(X_raw_n)


sklearn_w = np.concatenate([[model.intercept_], model.coef_])
mse_s = (1/(2*len(y_n))) * np.sum((pred_sk - y_n)**2)
print(f"Mean squarred error using sklearn: {mse_s}")
print(f"Gradient Descent weights: {weights_g}")
print(f"Analytical weights:       {weights_a}")
print(f"Sklearn weights:          {sklearn_w}")
print(f"Gradient Descent mean squarred error: {mse_g}")
print(f"Analytical mean squarred error: {mse_a}")
print(f"Sklearn mean squarred error: {mse_s}")
```
