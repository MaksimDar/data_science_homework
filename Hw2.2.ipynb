```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

1-Graph: Зв'язок між рівнем заробітної плати та загальним і конкретним досвідом роботи 

```python
fig,axes = plt.subplots(1,2,figsize=(10,4))
sns.barplot(x="exp",y="salary",data=developers,ax=axes[0]) 
sns.barplot(x="current_job_exp",y="salary",data=developers, ax=axes[1]) 
axes[0].set_title("Порівняння загального досвіду з рівнем доходу")
axes[1].set_title("Порівняння конкретного досвіду з рівнем доходів")
fig.suptitle("Зв'язок між рівнем заробітної плати та загальним і конкретним досвідом роботи",y=1.05)
plt.tight_layout()
plt.show()
```

2-Graph: Порівняння кількості співробітників за мовами програмування

```python
programming_languages = developers["Мова програмування"].unique().tolist()
```

```python
plt.bar(programming_languages,developers["Мова програмування"].value_counts(),color="#468432")
plt.xticks(rotation=90)
plt.xlabel("Мови програмування",size="medium",color="#406093")
plt.ylabel("Кількість співробітників",size="medium",color="#406093")
plt.title("Порівняння кількості співробітників за мовами програмування")
plt.show()
```

3-Graph: Порівняння освітнього рівня співробітників

```python
education_options = developers["Освіта"].unique().tolist()
plt.pie(developers["Освіта"].value_counts(),labels=education_options,autopct="%.2f%%")
plt.title("Порівняння освітнього рівня співробітників")
plt.show()
```

4-Graph: Розподіл заробітної плати за рівнем освіти та досвідом

```python
sns.stripplot(x="Освіта",y="salary",hue="exp",data=developers)
plt.xlabel("Типи наукових ступенів",size="medium",color='#FF8040')
plt.ylabel("діапазон заробітної плати",size="medium",color='#FF8040')
plt.title('Розподіл заробітної плати за рівнем освіти та досвідом',pad=20)
plt.xticks(rotation=45)
plt.show()
```

5-Graph: Залежність заробітної плати від рівня англійської мови та досвіду роботи

```python
sns.scatterplot(x="salary",y="Рівень англійської", hue="exp",data=developers)
plt.title("Залежність заробітної плати від рівня англійської мови та досвіду роботи")
plt.show()
```
